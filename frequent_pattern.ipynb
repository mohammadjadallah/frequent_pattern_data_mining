# Task3: Convert the data ( from task2) in the dataframe into a proper format using:  (iterrows() and append()).

```python
# we can use this method
# list(map(list, df.iterrows()))

# we can use this method to convert the data from df to list of list, so we can use the new data with mlxtend
import csv

with open(r'retail_dataset1.csv') as file:
    try:
        data = list(csv.reader(file))
    except:
        pass
data = data[1:]
```

```python
# Clean data from NaN values
new_data = []
for i in data:
    if '' in i:
        counts = i.count('')
        for j in range(counts):
            i.remove('')
        new_data.append(i)
```

```python
new_data
```

# Task4: Encode the  the previous dataset to dataframe, using Transactionencoder.

```python
import mlxtend.preprocessing as mlp

# Initialize the object
te = mlp.TransactionEncoder()

# Transform the data
tedata = te.fit(new_data).fit_transform(new_data)

ndf = pd.DataFrame(tedata, columns=te.columns_)
```

```python
ndf
```

# Task5: Use the Apriori algorithm to calculate the most frequent items and perform the associated-rule analysis on the given dataframe.  min_support = 0.3, use_colnames = True.

```python
import mlxtend.frequent_patterns as fp

# apriori method
fp_apr = fp.apriori(ndf, min_support=0.3, use_colnames=True)
fp_apr
```

# Task6: Use the fpgrowth algorithm to calculate the most frequent items and perform the associated-rule analysis on the given dataframe.  min_support = 0.3.

```python
# fpgrowth method

fpg = fp.fpgrowth(ndf, min_support=0.3, use_colnames=1)
```
